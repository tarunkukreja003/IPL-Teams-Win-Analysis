```python
df_win_loss_mi_clean_f = df_win_loss_mi_cleaned[:10]
df_win_loss_mi_clean_f = df_win_loss_mi_clean_f.replace(to_replace = '50%', value = '50.00%')
for i in np.arange(0,10,1):
    df_win_loss_mi_clean_f['Win%'][i] = df_win_loss_mi_clean_f['Win%'][i][:df_win_loss_mi_clean_f['Win%'][i].index('%')]
df_win_loss_mi_clean_f 
```
